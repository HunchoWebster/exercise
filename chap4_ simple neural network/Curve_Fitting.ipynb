```python

import tensorflow as tf

# 定义函数
def f(x):
  return x**3 + 0.5*x**2 - 2*x + 1

# 采样数据
x_train = tf.random.uniform([1000, 1], minval=-2, maxval=2)
y_train = f(x_train)
x_test = tf.random.uniform([10, 1], minval=-2, maxval=2)
y_test = f(x_test)
```

```python
# 定义模型
model = tf.keras.Sequential([
  tf.keras.layers.Dense(100, activation='relu'),
  tf.keras.layers.Dense(1)
])

# 编译模型
model.compile(optimizer='adam', loss='mse')

# 训练模型
model.fit(x_train, y_train, epochs=100)
```

```python
# 评估模型
loss = model.evaluate(x_test, y_test)
print('测试集误差：', loss)
```

```python




# 绘制图像
import matplotlib.pyplot as plt


plt.plot(x_test, y_test, 'r', label='Real')
plt.plot(x_test, model.predict(x_test), 'b', label='Predict')
plt.legend()
plt.show()
```
